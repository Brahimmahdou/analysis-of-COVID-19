```python
## features : indus, nox, rm, tax, lstat, exp_age
## target : medv
import pandas as pd
import numpy as np
df = pd.read_csv('C:\\Users\\Mahdou\\Downloads\\BostonHousing.csv')
for col_var in df.columns:
    column_to_check = 'col_var'
    Q1 = df[col_var].quantile(0.25)  ### calculate the first quartile Q1
    Q3 = df[col_var].quantile(0.75)  ### calculate the third quartile Q3
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col_var] < lower_bound) | (df[col_var] > upper_bound)] 
from sklearn.preprocessing import MinMaxScaler 
min_max_scal = MinMaxScaler()
out_index=outliers.index
features_scale = ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat', 'medv']
df = pd.read_csv('C:\\Users\\Mahdou\\Downloads\\BostonHousing.csv')
outliers_mask = df.index.isin(out_index)
clean_data = df[~outliers_mask]
scal_features = min_max_scal.fit_transform(clean_data[features_scale])
scal_df = pd.DataFrame(scal_features, columns=features_scale)

columns_to_delete = ['crim', 'zn', 'chas', 'dis', 'rad', 'ptratio', 'b']


scal_df.drop(columns=columns_to_delete, inplace=True)
df1=pd.DataFrame(scal_df)

#df1['log_lstat'] = np.log(df1['lstat'])
df1['exp_age'] = np.exp(df1['age'])
df1
```

```python
df1
```

```python
################################ 1) Model selection: ###############################################

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

# We define X as features and Y as target variable(medv)

X =  df1[['indus','nox','rm','tax','lstat','exp_age']]
Y = df1['medv']

# Split our data into (training sets) and (testing sets):
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42) # train and evaluate our model

# We implement baseline models: (dictionnary called models): 
models = {  ## (key represents the name of a machine learning model) (value represents the model object)
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Support Vector Machine": SVR(),
    "Neural Network": MLPRegressor()
}

# Train and evaluate baseline models:
for name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    print(f"The error of ({name}) is {mse}")

############### the best is Random forest 
```

```python
####################### 2) hyperparameter tuning : ##############################################
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor

# We define hyperparameter grid for Random Forest model:
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# We instantiate RandomizedSearchCV:
random_search = RandomizedSearchCV(RandomForestRegressor(), param_distributions=param_dist, n_iter=10, cv=5, random_state=42)
### RandomizedSearchCV: (class in scikit-learn used for hyperparameter tuning)

# We fit RandomizedSearchCV on training data:
random_search.fit(X_train, y_train)

# we get the best hyperparameters:
our_best_params = random_search.best_params_

# We train model with best hyperparameters:
our_best_model = RandomForestRegressor(**our_best_params)
our_best_model.fit(X_train, y_train)

# We valuate best model:
mse = mean_squared_error(y_test, our_best_model.predict(X_test))
print("Validation MSE for best model:", mse)
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestRegressor

# Instantiate the RandomForestRegressor model
#our_best_model = RandomForestRegressor(n_estimators=100, random_state=42)  # Example with 100 trees

# Define the k-fold Cross-Validation Strategy
k_fold = KFold(n_splits=5, shuffle=True, random_state=42)  # Example with 5 folds

# Perform Cross-Validation
scores = cross_val_score(our_best_model, X, Y, cv=k_fold, scoring='neg_mean_squared_error')

# Evaluate Cross-Validation Results
mean_score = scores.mean()
std_score = scores.std()

# Print the mean and standard deviation of the cross-validation scores
print("Mean Cross-Validation Score:", mean_score)
print("Standard Deviation of Cross-Validation Score:", std_score)
```
